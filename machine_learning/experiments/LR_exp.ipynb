For this notebook we will focus on Linear Regression and verify the robustness of the model made from scratch

```python
import sys
import os
project_root = os.path.abspath(os.path.join(os.getcwd(), "..", ".."))

if project_root not in sys.path:
    sys.path.append(project_root)

print(project_root)
```

```python
from machine_learning.models.linear_regression import LinearRegression
```

```python
# Dataset
X = [0, 1, 2, 3]
y = [1, 3, 5, 7] 

# y = 1 + 2*x

model = LinearRegression()

model.fit(X, y)
```

```python
# Check the beta_0, and beta_1 
# Should be beta_0 = 1, beta_1 = 2
print(model.beta_0)
print(model.beta_1)
```

```python
y_pred = model.predict(X)
print("y_pred:", y_pred)
print("y_true:", y)
```
